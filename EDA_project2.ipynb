```python
sns.countplot(x='school',data=df)
plt.title('SCHOOL',color="b")
plt.show()
```

```python
sns.histplot(x='famsize',data=df)
plt.title('FAMILY SIZE',color="red")
plt.show()
```

```python
sns.boxplot(x='age',data=df,color="blue")
plt.title('AGE',color="green")
plt.show()
```

```python
numdf=df.select_dtypes(include=np.number)
plt.figure(figsize=(15,10))
sns.heatmap(numdf.corr(),annot=True)
plt.title("Heatmap",color="blue")
plt.show()
```

```python
sns.scatterplot(x='absences',y='G3',data=df,color="green",marker="*")
plt.title('ABSENCES VS G3',color="red")
plt.show()
```

```python
sns.lineplot(x='absences',y='G3',data=df,color="green")
plt.title('ABSENCES & G3',color="red")
plt.show()
```
